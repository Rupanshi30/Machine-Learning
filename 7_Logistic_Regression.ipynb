#### Dropping irrelevant columns

```python
df.drop(columns=["id","Unnamed: 32"],inplace=True)
```

- Since target is categorical, so we will encode it using label encoding as
- m:1 and b:0

```python
df["diagnosis"]=df["diagnosis"].replace({"M":1,"B":0})
df
```

#### Seggregate the features and target

```python
x=df.drop(columns=["diagnosis"])
y=df["diagnosis"]
```

- Different units of columns, that's why we are doing scaling of features.
- scaling is required in linear and logistic

```python
sc=StandardScaler()
x_scaled=sc.fit_transform(x)
```

```python
x_scaled
```

#### Split the data for training and model evaluation

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=120)
```

#### checking the shape of train and test data

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_test.shape)
print(y_test.shape)
```

#### Model Creation

- Model of logistic Regression

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

#### Model Evaluation

```python
y_pred=lr.predict(x_test)
```

#### Checking for imbalanced data but it does not matter to logistic, it is robust.

#### We will not rely on accuracy if imbalanced data.

```python
df["diagnosis"].value_counts()
```

- Not imbalanced Data
- Can rely on accuracy for this model.
- But need to check error i.e. confusion matrix for this data as this case(Breast Cancer) is sensitive.

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
true_positive=cm[0,0]
true_negaitive=cm[1,1]
false_positive=cm[1,0]
false_negative=cm[0,1]
```

#### Accuracy on train Data

```python
accuracy_score(y_test,y_pred)
```

- 97% accuracy as error is less but check for train accuracy too.

- Since the accuracy on test data is so high so there is a chance of overfitting, so we check the accuracy on training data.

#### Accuracy on test Data

```python
y_pred_train=lr.predict(x_train)
```

```python
accuracy_score(y_train,y_pred_train)
```

- From this output, we can say that the variance is less so this is not the case of overfitting.

#### For this case, we will focus on Type-II error, so recall, no need to check precision

```python
recall=true_positive/(true_positive+false_negative)
recall
```

```python
f1_score(y_test,y_pred)
```

- Not much difference in accuracy and f1_score, so no overfitting.
